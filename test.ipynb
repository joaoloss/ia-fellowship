```python
import spacy

nlp = spacy.load("xx_ent_wiki_sm")  # Multilingual NER model
text = "son goku"

doc = nlp(text)

for ent in doc.ents:
    if ent.label_ == "PER":
        print("Detected name:", ent.text)
```

```python
a = "example string one"
b = "example string two"

dist = editdistance.eval(a, b)
longest = max(len(a), len(b))
print(dist, longest)
print(dist / longest < 0.20)
```
