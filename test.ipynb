```python
from datetime import datetime

def is_date(value: str) -> bool:
    """
    Check if the value is a date in common formats.
    """
    date_formats = [
        "%Y-%m-%d",
        "%d/%m/%Y",
        "%m/%d/%Y",
        "%d-%m-%Y",
        "%m-%d-%Y",
        "%d.%m.%Y",
        "%m.%d.%Y",
    ]
    for fmt in date_formats:
        try:
            datetime.strptime(value, fmt)
            return True
        except ValueError:
            continue
    return False

print(is_date("2023-10-15"))  # Expected: True
print(is_date("15/10/2023"))  # Expected: True
print(is_date("10.15.2023-90"))  # Expected: True
print(is_date("Hello World"))  # Expected: False
```
